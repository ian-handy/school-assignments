```python
import pandas as pd
import numpy as np

df = pd.read_csv('grade_record.csv')

df.rename(columns={'Points':'Grade','Grade':'Points'},inplace=True) #Fix incorrect column names because it was bugging me

def find_fixed_grade(grade_sheet):
    points = grade_sheet['Points']
    grade_ranges = [
        points >=90,
        points >=80,
        points >=70,
        points >=60
    ]
    grade = ['A','B','C','D']
    grade_sheet['Fixed Grade'] = np.select(grade_ranges,grade,default='F')
    return grade_sheet

find_fixed_grade(df)

incorrect_grades = [df['Grade'][i] for i in range(len(df)) if not df['Grade'][i]==df['Fixed Grade'][i]]
print(f'{len(incorrect_grades)} incorrect grades.\n{100 - round((len(incorrect_grades)/len(df)*100))}% accuracy.')

```

```python
df.pop('Grade')
df.rename(columns={'Fixed Grade':'Grade'},inplace=True)
df.to_csv('fixed_grade_record.csv', index=False)
```
