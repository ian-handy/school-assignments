```python
import pandas as pd

df = pd.read_csv('datasets/TitanicSurvival.csv')

def class_survived(class_name):
    class_total = df[df['passengerClass']==class_name]
    survived = class_total['survived'].value_counts().get('yes')
    print(f'{(survived/len(class_total)*100):.2f}% of {class_name} class passengers survived.')

class_survived('1st')
class_survived('2nd')
class_survived('3rd')
```
