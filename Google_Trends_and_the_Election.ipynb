#**Google Trends and the Election**



Before I begin, I would like to make it clear that there is almost certainly a non-zero chance that, because Kamala Harris is referred to as both "Kamala" and "Harris", there will be a large portion of searches that I am not including if I simply use either her first or last name. I will use the candidates last names in the second and third graphs for the sake of simplicity and time efficiency, knowing full well that I am not including the full scope.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('trump-harris-election.csv', header=1)
df['Day'] = pd.to_datetime(df['Day'])

df.plot(x = "Day").set_ylabel("Interest over time")
plt.show()
```

This information (sourced from Google Trends) shows the relative search interest on Google between Donald Trump, Kamala Harris, and the election. The first date is July 20th, 2024, and the last date is October 31st, 2024.

I chose the starting point as one day before July 21st, 2024, the day when Joe Biden withdrew from the 2024 Presidential Election and announced that he was endorsing Kamala Harris.

The ending date was originally going to be the day of the election, but, due to massive spikes in searches for "election" this skewed the relativity of the rest of the data enormously and left very little variation between the dates for the rest of the year. Hence, I chose the last day of October as a date that did not take away from the spike of searches on July 21st, 2024.

```python
import matplotlib.pyplot as plt

df = pd.read_csv('debate.csv', header=1)
df['Day'] = pd.to_datetime(df['Day'])

for column in ['trump debate: (United States)', 'harris debate: (United States)', 'biden debate: (United States)']:
    df[column] = pd.to_numeric(df[column], errors='coerce')

df.plot(x = "Day").set_ylabel("Interest over time")
plt.show()
```

The above graph represents country-wide searches for the debates between Joe Biden and Donald Trump and between Kamala Harris and Donald Trump. It is interesting to note here that there is a higher search rate for the candidate who "lost" the debate according to polls conducted after the debates were over.

*   ["Majority of Americans say Harris won the debate"](https://www.ipsos.com/en-us/abc-news-ipsos-september-post-debate-2024)



```python
import matplotlib.pyplot as plt

df = pd.read_csv('debate-win.csv', header=1)
df['Day'] = pd.to_datetime(df['Day'])

df.plot(x = "Day").set_ylabel("Interest over time")
plt.show()
```

The above graph shows searches for "(president) win debate" where president is "trump", "harris", or "biden". Interestingly, there were relatively more searches of "win debate" for results concerning President Donald Trump for both debates, even though the previous graph shows more relative interest for searches of "Biden" in relation to the debate than for "Trump".

```python
import matplotlib.pyplot as plt

df = pd.read_csv('kamala-harris.csv', header=1)
df['Day'] = pd.to_datetime(df['Day'])

for column in ['kamala: (United States)', 'harris: (United States)', 'kamala harris: (United States)']:
    df[column] = pd.to_numeric(df[column], errors='coerce')

df.plot(x = "Day").set_ylabel("Interest over time")
plt.show()
```

The above graphs show an interesting relationship between Kamala Harris, Donald Trump, and Kamala Harris and how people are searching for them on Google.

Joe Biden and Donald Trump have relatively low search results for their first names  and full names as opposed to only their last names, while Kamala Harris, though still most commonly searched for by her last name, has a much stronger presence of searches for her first name alone and her full name.
