```python
import numpy as np
import os
import pandas as pd
from urllib.request import urlretrieve


url = "https://www.nhc.noaa.gov/data/hurdat/hurdat2-1851-2017-050118.txt"

local_fname = "hurdat2.txt"
if not os.path.exists("hurdat2.txt"):
    urlretrieve(url, local_fname)
```

```python
colnames = ['date','time','record_id','status','latitude','longitude','max_wind','min_pressure','ne34ktr','se34ktr','sw34ktr','nw34ktr','ne50ktr','se50ktr','sw50ktr','nw50ktr','ne64ktr','se64ktr','sw64ktr','nw64ktr']
df = pd.read_csv("./hurdat2.txt",names=colnames,skipinitialspace=True,usecols=range(20))
df
```

```python
test_df = df[df['date'].str.contains(r'[A-Z]')]
test_df.sort_values('date')
test_df
```

```python
id_dic = {'date':'identifier','time':'name','record_id':'num_pts'}
id_df = df[df['date'].str.contains('AL')][id_dic.keys()].rename(id_dic,axis=1)
id_df
```

```python
df.drop(index=id_df.index,inplace=True)
id_df.index+=1
df
```

```python
df = id_df.join(df,how='outer').reset_index(drop=True)
for col in id_dic.values():
    df[col] = df[col].ffill()
df
```

```python
df.replace([-999,'UNNAMED'],np.nan,inplace=True)
df['date']=pd.to_datetime((df['date']+df['time']))
df = df.drop('time',axis=1).rename({'date':'datetime'},axis=1)
df
```

```python
def convert_coord(x):
    x = str(x)
    num = float(x[:-1])
    if x[-1] in 'SW':
        return -num
    else:
        return num

df[['latitude','longitude']] = df[['latitude','longitude']].map(convert_coord)
df
```

```python
df.to_csv('hurdat2.csv',index=False)
df = pd.read_csv('./hurdat2.csv')
df
```

```python
import altair as alt
import vega_datasets

states_data = alt.topo_feature(vega_datasets.data.us_10m.url, "states")
states = alt.Chart(states_data).mark_geoshape(
    stroke='black',
    strokeWidth=1
    ).transform_filter((alt.datum.id != 2) & (alt.datum.id != 15))


df = pd.read_csv('./hurdat2.csv', parse_dates=['datetime'])

maria = df[(df['datetime'] > '2017-01-01') & (df['name'] == 'MARIA')]

hurricanes = alt.Chart(maria).mark_line().encode(
    longitude='longitude:Q',
    latitude='latitude:Q',
    order='datetime:T'
)

(states + hurricanes).properties(width=700, height=400, title="Hurricane Maria (2017)").interactive()
```
