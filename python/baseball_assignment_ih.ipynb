```python
import pandas as pd
df_original = pd.read_excel("datasets/kaggle_data_moneyball_baseball_1999_2012.xlsx",header=1)
df = df_original
display(df)
```

```python
df.corr(numeric_only=True)
```

In the interest of getting a good price, we are going to exclude any teams who have been in the Playoffs

```python
df.pop('G') #All the same
df.drop(df[df['Playoffs'] == 1].index,inplace=True)
df.drop(df[df['RankPlayoffs'] == 1].index,inplace=True)
display(df)
```

RA,OOBP, and OSLG have the strongest negative correlation for wins, so we are going to purge the teams with valeus lower than the 3rd quartile

```python
ra_mean = df['RA'].quantile(.75)
oobp_mean = df['OOBP'].quantile(.75)
oslg_mean = df['OSLG'].quantile(.75)
print(f'RA 3rd Quartile: {ra_mean}\n OOBP 3rd Quartile: {oobp_mean}\n OSlG 3rd Quartile: {oslg_mean}')
```

```python
df.drop(df[df['RA'] <= ra_mean].index,inplace=True)
df.drop(df[df['OOBP'] <= oobp_mean].index,inplace=True)
df.drop(df[df['OSLG'] <= oslg_mean].index,inplace=True)
print(df.shape)
display(df)
```

```python
df.corr(numeric_only=True)
```

Wins have a high correlation with OBP for the remaining teams, so we are going to prioritize teams with high values in this area

```python
df.sort_values('OBP',ascending=False,inplace=True)
display(df)
```

Teams with high winrates are going to (presumably) fetch higher prices, so we are now going to purge teams with exceptionally high wins

```python
win_mean = df['W'].quantile(.75)
df.drop(df[df['W'] >= win_mean].index,inplace=True)
display(df)
```

The teams we are left with are seemingly unexceptional, but the teams at the top have the potential to shine given their having been saved from many rounds of culling.
