```python
import pandas as pd
import altair as alt

df = pd.read_csv("./hurdat2-1.csv")
df.head()
```

```python
df_names = df[["name","identifier"]].drop_duplicates()
df_names = df_names[df_names["name"] != "UNNAMED"]
df_names
```

```python
pd.DataFrame(df_names["name"].value_counts()).iloc[0]
```

```python
df_names["year"] = df_names['identifier'].str[-4:]
pd.DataFrame(df_names.value_counts('year')).iloc[0]
```

```python
df.iloc[df["latitude"].idxmax()]
```

```python
df[df["min_pressure"]!=-999].sort_values('min_pressure').iloc[0]
```

```python
df["year"] = df["datetime"].str[:4]
df["month"] = df["datetime"].str[5:7]
ct_df = df.drop_duplicates(subset='identifier')
```

```python
with pd.option_context('display.max_rows', 10):
    ct = pd.crosstab(ct_df["year"],ct_df["month"],margins=True)
    display(ct)
```

```python
df_maria = df[(df['name'] == 'MARIA') & (df['year'] == '2017')]
alt.Chart(df_maria).mark_line(point=True).encode(
    alt.X('min_pressure', scale=alt.Scale(zero=False)),
    alt.Y('max_wind', scale=alt.Scale(zero=False)),
    order='datetime'   
)
```
