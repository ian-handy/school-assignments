```python
import os
from urllib.request import urlretrieve

# download the data if we don't have it locally
url = "https://www.nhc.noaa.gov/data/hurdat/hurdat2-1851-2017-050118.txt"
local_fname = "hurdat2.txt"
if not os.path.exists("hurdat2.txt"):
    urlretrieve(url, local_fname)

# very primtive way of reading the data
# can be improved
records = []
with open(local_fname,'r') as f:
    for line in f:
        if line.startswith("AL"):
            record = line.strip()
            reports = []
            records.append((record, reports))
        else:
            reports.append(line.strip())
records
```

Each record is a tuple that consists of
0. a string with the hurricane info 
1. a list of track points where each point is as documented [here](https://www.nhc.noaa.gov/data/hurdat/hurdat2-format-atlantic.pdf)

```python
# do your work for Problem 1 here
unique_names = {}
for record in records:
    name = record[0].split(',')[1].strip()
    if not name == 'UNNAMED':
        if not name in unique_names:
            unique_names.update({name : 1})
        else:
            count = unique_names[name] + 1
            unique_names.update({name:count})
print(unique_names)
len(unique_names)
```

## Problem 4

```python
# do your work for Problem 4 here
highest_latitude = 0
name = ''
year = 0
for record in records:
    current_latitude = max([float(lat.split(',')[4][0:-1]) for lat in record[1]])
    if current_latitude > highest_latitude:
        highest_latitude = current_latitude
        name = record[0].split(',')[1].strip()
        year = int(record[1][0].split(',')[0][0:4])
print((name,year))
```
