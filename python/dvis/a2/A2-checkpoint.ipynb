2-a: Yearly Bar Chart

We will help you walk through the first question, which may give you an impression of how Altair works.
We basically need to indicate:
1. data name: "data"
2. mark type: "mark_bar()"
3. encode: x, y, color, etc...

Now, let's check the code line by line:
1. "alt.Chart(data).mark_bar().encode()" indicate the data file and mark type. In "encode()", we will specify the data in x and y dimension and their colors.
2. "year(data):O" will return the year and O indicates that this is an ordinal attribute.
3. "distinct()" will return unique hurricane ID, and count() will count the items (hurricanes) of each year.
4. "month(datetime):O" will return the month.

```python
alt.Chart(data).mark_bar().encode(
    alt.X('year(datetime):O'),
    alt.Y('count(identifier):Q', aggregate = 'distinct'),
    color='month(datetime):O'
)
```

2-b: Scatterplot of Wind and Pressure

There are two challenges: 1. exclude zero from min_pressure; 2. exclude -999 from min_pressure. The following code segment may help you. 

Note you need to replace "xxx" by correct contents. 


```python
alt.Chart(data).xxx().encode(
    alt.X('xxx', axis = alt.Axis(title='Max Wind')),
    alt.Y('xxx', scale=alt.Scale(zero=False), axis = alt.Axis(title='Min Pressure'))
).transform_filter(alt.datum.min_pressure>0)
```
